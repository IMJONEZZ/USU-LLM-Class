```python
# compare_tokenizers.ipynb

import json
from Tokenizer_Smiple import SimpleTokenizer, build_vocab
from Tokenizer_BPE import BPE
```

```python

# Load JSON file
with open("SW_EpisodeIV_VI.json", "r", encoding="utf-8") as f:
    data = json.load(f)

# Get a sample line from the script
sample_text = data[0]["Line"]
print(f"Original Line: {sample_text}")

```

```python
# ---- Simple Tokenizer ----
# Build vocab
simple_vocab = build_vocab([entry["Line"] for entry in data])
simple_tokenizer = SimpleTokenizer(simple_vocab)

# Tokenize with SimpleTokenizer
simple_tokens = simple_tokenizer.encode(sample_text)
simple_decoded = simple_tokenizer.decode(simple_tokens)

print("\nSimple Tokenizer:")
print(f"Tokens: {simple_tokens}")
print(f"Decoded Text: {simple_decoded}")


```

```python
# ---- BPE Tokenizer ----
# Train BPE tokenizer
bpe_tokenizer = BPE([entry["Line"] for entry in data], vocab_size = 1000)
bpe_tokenizer.train()

# Tokenize with BPE
bpe_tokens = bpe_tokenizer.tokenize(sample_text)

print("\nBPE Tokenizer:")
print(f"Tokens: {bpe_tokens}")
```
