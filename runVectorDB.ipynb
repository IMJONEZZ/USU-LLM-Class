```python
import os
import shutil
import subprocess

# Set repository details
REPO_URL = "https://github.com/IMJONEZZ/USU-LLM-Class.git"
BRANCH_NAME = "matt-burns-assignment-7"

# Get repository name from URL
repo_name = REPO_URL.split("/")[-1].replace(".git", "")

# Check if directory exists and remove it
if os.path.exists(repo_name):
    print(f"Repository {repo_name} exists. Removing...")
    shutil.rmtree(repo_name)
    print(f"Repository {repo_name} removed successfully.")

# Clone the repository with the specific branch
print(f"Cloning repository {REPO_URL} (branch: {BRANCH_NAME})...")
subprocess.run(["git", "clone", "-b", BRANCH_NAME, REPO_URL])
print(f"Repository cloned successfully.")

# Change directory to the repository
print(f"Changing directory to {repo_name}...")
os.chdir(repo_name)
print(f"Current directory: {os.getcwd()}")
```
