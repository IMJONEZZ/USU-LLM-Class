````python
from finetuning_custom_data import data as QandA_data
from typing import List
from pydantic import BaseModel, Field
from distilabel.pipeline import Pipeline
from distilabel.steps import LoadDataFromDicts
from distilabel.steps.tasks import TextGeneration

class Q_and_A(BaseModel):
    question: str = Field(..., description="The question to be answered")
    answer: str = Field(..., description="The correct answer to the question")

class Q_and_As(BaseModel):  
    exam: List[Q_and_A]


SYSTEM_PROMPT = """\
You are an expert in creating questions and answers for teaching LLMs to reason effectively.
Your goal is to create questions and answers that test similar skills as the example questions and answers.
Your answer must adhere to the following format:
```
[
    {
        "question": "Your question",
        "answer": "The correct answer to the question",
    },
    ... (more questions and answers as required)
]
```
""".strip()

def remap_keys(example: dict) -> dict:
    example["instruction"] = f"Question: {example['question']}\nAnswer: {example['answer']}"
    return example

processed_data = [remap_keys(item) for item in QandA_data]

with Pipeline(name="QandAGenerator") as pipeline:

    load_dataset = LoadDataFromDicts(
        name="load_instructions",
        data=processed_data
    )
    
    text_generation = TextGeneration(  
        name="QandA_generation",
        system_prompt=SYSTEM_PROMPT,
        template="Generate a list of questions and answers similar to the following. {{instruction}}",
        llm=OllamaLLM(
            model="llama3.3",
            structured_output={
                "schema": Q_and_As.model_json_schema(),
                "format": "json"
            }
        ),
        input_batch_size=8,
        output_mappings={"model_name": "generation_model"},
    )

    load_dataset >> text_generation
    
    distiset = pipeline.run()
````
